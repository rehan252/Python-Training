```python
res = requests.get("http://www.pakpositions.com/search-jobs/python")
```

```python
s = BeautifulSoup(res.text, 'html.parser')
```

```python
jobs = s.find_all('div', {'class' : 'intlist'})
```

```python
import csv

jobfile = csv.writer(open('jobs2.csv','a'))
jobb = []
desc = []
for job in jobs:
    job_title = job.find('a', {'class' : 'jobtitle'}).text
    #company = job.find('div', {'class' : 'cname'})
    jobb.append(job_title)
    des = job.find('p').text
    des = des.replace("\n"," ")
    des = des.replace("\r"," ")
    des = des.replace('\xa0',"")
    desc.append(des)

jobfile.writerow(jobb)
jobfile.writerow(desc)
```

```python
for job in jobs:
    print(job.find('p').text)
    print('\n-------------')
```
